```python
import requests
# headers
headers = {
    "User-Agent":'Mozilla/6.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'
    }

# send request to download the data
resp = requests.request("GET", 'https://www.amazon.in/s?k=one+plus+7&dc&ref=a9_asc_1', headers=headers)
resp.status_code, resp.url
```

```python
data = resp.text

pprint(data)
```

```python
soup = BeautifulSoup(data)
print(soup.prettify())
```

```python
for i in soup.find_all('img'):
    print(i['src'])
```

```python
for i in soup.find_all("a",  attrs = {'href': re.compile("^http")}):
    print(i['href'])
```

```python
for i in soup.find_all(class_ = 'a-price-whole'):
    print(i.text)
```

```python
prices = []
for i in soup.find_all(class_ = 'a-price-whole'):
    prices.append(i.text)
```

```python
prices
```

```python
len(prices)
```

```python
for i in soup.find_all(class_ ="a-size-medium a-color-base a-text-normal")[3:]:
    print(i.text)
```

```python
names = []
for i in soup.find_all(class_ ="a-size-medium a-color-base a-text-normal")[2:]:
    names.append(i.text)
```

```python
len(names)
```

```python
import pandas as pd
amazon = pd.DataFrame()
amazon['names'] = names
amazon['prices'] = prices
```
