```python
resp = requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area')
```

```python
soup = BeautifulSoup(resp.text, "html.parser")

soup.prettify()
```

#### Extracting table using attribute "class" and it svalue is "wikitable sortable"

```python
table_data = soup.table

table_data
```

```python
table_data = soup.find(class_ = "wikitable sortable")

table_data
```

#### Extracting all the links from a table

```python
table_links = table_data.find_all("a")
```

```python
table_links
```

```python
countries = []
```

```python
for link in table_links:
    countries.append(link.get("title"))
```

```python
countries
```

```python
countries.remove(None)

countries
```

```python
table_headings = table_data.find_all('th') 

table_headings
```

#### Extracting table headings

```python
for th in table_headings:
    print(th.text)
```
